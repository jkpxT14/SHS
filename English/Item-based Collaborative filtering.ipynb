## [Item-based Collaborative filtering]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
```

```python
rating_data=pd.read_csv("Movielens/ratings.csv")

print(rating_data)
```

```python
movie_data=pd.read_csv("Movielens/movies.csv")

print(movie_data)
```

```python
rating_data.drop("timestamp", axis=1, inplace=True)

print(rating_data)
```

```python
movie_user_rating=pd.merge(rating_data, movie_data, on="movieId")

print(movie_user_rating)
```

```python
movie_user_rating=movie_user_rating.pivot_table("rating", index="title", columns="userId")

print(movie_user_rating)
```

```python
movie_user_rating.fillna(0, inplace=True)

print(movie_user_rating)
```

```python
item_based_collabor=cosine_similarity(movie_user_rating)
print(item_based_collabor)
```

```python
item_based_collabor=pd.DataFrame(data=item_based_collabor, index=movie_user_rating.index, columns=movie_user_rating.index)
print(item_based_collabor)
```

```python
def get_item_based_collabor(title):
    return item_based_collabor[title].sort_values(ascending=False)[1:11]
```
