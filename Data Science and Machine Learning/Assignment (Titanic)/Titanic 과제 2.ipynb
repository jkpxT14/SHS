```python
ax=train["Survived"].value_counts().plot.pie(autopct='%.2f')
ax.set_title("Survival Rate")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
# print(train[["Sex", "Survived"]])
train[["Sex", "Survived"]].groupby(["Sex"]).mean().plot.bar(ax=ax[0])
# ax[0].set_title("")
sns.countplot(train, x="Sex", hue="Survived", ax=ax[1])
# ax[1].set_title("")
plt.show()
```

```python
Survived=[136, 87, 119]
Not_Survived=[80, 97, 372]
Pclass=["1", "2", "3"]
plt.bar(Pclass, Survived, color='green', label="Survived")
plt.bar(Pclass, Not_Survived, color='red', label="Not Survived")
plt.legend(loc='lower left', bbox_to_anchor=(0.7, 1.0))
plt.show()
```

```python
Male_Survived=[45, 17, 47]
Male_Not_Survived=[77, 91, 300]
Female_Survived=[91, 70, 72]
Female_Not_Survived=[3, 6, 72]

plt.subplot(1, 2, 1)
plt.bar(Pclass, Male_Survived, color='green', label="Survived")
plt.bar(Pclass, Male_Not_Survived, color='red', bottom=np.array(Male_Survived), label="Not Survived")
plt.title("Male")

plt.subplot(1, 2, 2)
plt.bar(Pclass, Female_Survived, color='green', label="Survived")
plt.bar(Pclass, Female_Not_Survived, color='red', bottom=np.array(Female_Survived), label="Not Survived")
plt.title("Female")

plt.tight_layout()
plt.show()
```
