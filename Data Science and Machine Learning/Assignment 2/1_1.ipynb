```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([[74], [81], [94], [68], [64], [70], [83], [94], [71], [74]]) # Mock
y=np.array([71, 78, 90, 68, 66, 72, 89, 92, 78, 80]) # Final
```

```python
plt.scatter(x, y, color='blue')
plt.xlabel("Mock")
plt.ylabel("Final")
plt.show()
```

```python
print(np.corrcoef(x.reshape(len(y)), y))
```

```python
linReg=LinearRegression(fit_intercept=True)
model=linReg.fit(x, y)

w=model.coef_
b=model.intercept_
print(w, b, model.score(x, y))
```

```python
plt.plot(x, y, 'bo', label="Actual")
plt.plot(x, np.array(w*x+b), 'r', label="Predicted")
plt.xlabel("Mock")
plt.ylabel("Final")
plt.legend()
plt.show()
```

```python
xNew=np.array([[80], [90]])
yNew=model.predict(xNew)
print(yNew)
```
