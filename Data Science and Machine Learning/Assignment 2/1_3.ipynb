```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
x=np.array([[74], [81], [94], [68], [64], [70], [83], [94], [71], [74]])
y=np.array([1, 1, 1, 0, 0, 0, 1, 1, 0, 1])
```

```python
logReg=LogisticRegression(solver='lbfgs')
model=logReg.fit(x, y)

w=model.coef_
b=model.intercept_
print(w, b)
```

```python
def H(x):
    return 1/(1+np.exp(-(w*x+b)))
```

```python
yPred=np.where(H(x)>=0.5, 1, 0)
# yPred=model.predict(x)
# print(yPred)
```

```python
xIn=np.arange(x.min(), x.max(), 0.5)
Hx=H(xIn).reshape(-1, 1)

plt.plot(x, y, 'bo', label="Actual")
plt.plot(xIn, Hx, 'g', label="H(x)")
plt.plot(x, yPred, 'rx', label="Predicted")
plt.xlabel("Score")
plt.ylabel("Result")
plt.legend(loc='right')
plt.show()
```
