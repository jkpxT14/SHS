```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/greatsong/2023dataset/main/seoultemp(year2022).csv")
df=df.fillna(method='bfill')
print(df)
x=df['년'].values/1000
y=df['평균기온(℃)'].values
plt.scatter(x, y)
```
