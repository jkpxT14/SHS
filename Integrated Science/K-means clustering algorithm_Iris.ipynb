## <수능특강 - Unsupervised Learning - Clustering - K-means clustering algorithm>

#### 수능특강 독서 P. 170~173

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
iris=load_iris()

print(iris)
```

```python
iris_df=pd.DataFrame(data=iris.data, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])

print(iris_df)
```

#### 수능특강 독서 P. 172 2번 문제 관련

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(iris_df)
```

```python
iris_df["target"]=iris.target

print(iris_df["target"])
```

```python
iris_df["cluster"]=kmeans.labels_

print(iris_df["cluster"])
```

```python
# [Study] groupby
iris_result=iris_df.groupby(["target", "cluster"])["sepal_length"].count()

print(iris_result)
```

```python
pca=PCA(n_components=2)
pca_transformed=pca.fit_transform(iris.data)

print(pca_transformed)
```

```python
iris_df["pca_1"]=pca_transformed[:, 0]
iris_df["pca_2"]=pca_transformed[:, 1]

print(iris_df)
```

```python
idx_0=iris_df[iris_df["cluster"]==0].index
idx_1=iris_df[iris_df["cluster"]==1].index
idx_2=iris_df[iris_df["cluster"]==2].index

print(idx_0)
print(idx_1)
print(idx_2)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].scatter(iris_df.loc[:50, "pca_1"], iris_df.loc[:50, "pca_2"], marker='o', color='red')
ax[0].scatter(iris_df.loc[50:100, "pca_1"], iris_df.loc[50:100, "pca_2"], marker='^', color='green')
ax[0].scatter(iris_df.loc[100:, "pca_1"], iris_df.loc[100:, "pca_2"], marker='s', color='blue')
ax[0].set_xlabel("PCA 1")
ax[0].set_ylabel("PCA 2")

ax[1].scatter(iris_df.loc[idx_0, "pca_1"], iris_df.loc[idx_0, "pca_2"], marker='o', color='red')
ax[1].scatter(iris_df.loc[idx_1, "pca_1"], iris_df.loc[idx_1, "pca_2"], marker='^', color='green')
ax[1].scatter(iris_df.loc[idx_2, "pca_1"], iris_df.loc[idx_2, "pca_2"], marker='s', color='blue')
ax[1].set_xlabel("PCA 1")
ax[1].set_ylabel("PCA 2")

plt.show()
```

```python
centers=pd.DataFrame(kmeans.cluster_centers_, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])

print(centers)
```
